## Compare Spread/Total Sides vs. Combining the two in Same Game Parlay (Odds from DraftKings)

```python
from functions_sgp.gambling_odds import cnvt_deci_to_prob
```

```python
road_team = 'Dallas'
home_team = 'Las Vegas'
home_team_spread = 3.5
game_total = 50.5
road_team_spread_decimal_odds = 1.92
home_team_spread_decimal_odds = 1.89
total_over_decimal_odds = 1.92
total_under_decimal_odds = 1.89
road_team_spread_over_total = 3.64
road_team_spread_under_total = 3.54
home_team_spread_over_total = 3.60
home_team_spread_under_total = 3.42
```

```python
road_team_spread_prob = cnvt_deci_to_prob(road_team_spread_decimal_odds)
home_team_spread_prob = cnvt_deci_to_prob(home_team_spread_decimal_odds)

total_over_prob = cnvt_deci_to_prob(total_over_decimal_odds)
total_under_prob = cnvt_deci_to_prob(total_under_decimal_odds)

road_team_spread_over_house_odds = cnvt_deci_to_prob(road_team_spread_over_total)
road_team_spread_over_calc_ind_odds = road_team_spread_prob * total_over_prob
road_team_spread_under_house_odds = cnvt_deci_to_prob(road_team_spread_under_total)
road_team_spread_under_calc_ind_odds = road_team_spread_prob * total_under_prob

home_team_spread_over_house_odds = cnvt_deci_to_prob(home_team_spread_over_total)
home_team_spread_over_calc_ind_odds = home_team_spread_prob * total_over_prob
home_team_spread_under_house_odds = cnvt_deci_to_prob(home_team_spread_under_total)
home_team_spread_under_calc_ind_odds = home_team_spread_prob * total_under_prob
```

```python
print(f"""
Comparing "Independent Event" Calculation of SGP vs. "House Odds" Calculation of SGP
      
{road_team} Prob to Beat Spread {home_team_spread*-1}: {round(road_team_spread_prob*100, 2)}%
{home_team} Prob to Beat Spread {home_team_spread}: {round(home_team_spread_prob*100, 2)}%
Odds for Total to Go Over {game_total}: {round(total_over_prob*100, 2)}%
Odds for Total to Go Under {game_total}: {round(total_under_prob*100, 2)}%
{road_team} Beat Spread Total Over: Calculated off Leg Odds - {round(road_team_spread_over_calc_ind_odds*100, 2)}% vs. House Price - {round(road_team_spread_over_house_odds*100, 2)}%
{road_team} Beat Spread Total Under: Calculated off Leg Odds - {round(road_team_spread_under_calc_ind_odds*100, 2)}% vs. House Price - {round(road_team_spread_under_house_odds*100, 2)}%
{home_team} Beat Spread Total Over: Calculated off Leg Odds - {round(home_team_spread_over_calc_ind_odds*100, 2)}% vs. House Price - {round(home_team_spread_over_house_odds*100, 2)}%
{home_team} Beat Spread Total Under: Calculated off Leg Odds - {round(home_team_spread_under_calc_ind_odds*100, 2)}% vs. House Price - {round(home_team_spread_under_house_odds*100, 2)}%
""")
```
