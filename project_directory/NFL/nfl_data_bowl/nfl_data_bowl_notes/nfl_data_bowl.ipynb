# NFL Big Data Bowl 2025
## Two Pools
1. Analytics: Develop a Model to Study Player Movement After Ball is Thrown
    - Focus on the TARGETED player on offense and MULTIPLE defensive players.
    - Isolate to when the ball is IN THE AIR so post-catch metrics not relevant here.
        - Change of Direction for defensive player
        - Acceleration of Receiver
        - Defensive Scheme and How Multiple Defenders Relate
        - Closing Burst of Defender
    - Requirements:
        - 2000 word (max) writeup
        - Media Gallery (up to 10 images/videos/tables/figures)
        - Public Video on YouTube (3 minutes)
2. Prediction: Create a Model to Predict Player Movement After Ball is Thrown

## NFL Analytics Train Set
- 23 Total Columns
    - game_id, play_id, play_direction, absolute_yardline_number, num_frames_output
    - player_to_predict (True/False)
    - nfl_id (player_id)
        - player_name, player_height, player_weight, player_birth_date, player_position, player_side (offense/defense)
        - player_role
            - defensive_coverage, passer, targeted_receiver, other_route_runner
    - frame_id
        - player_specific
            - x: player position on "length of field" (0-120 yards, think the field from the press box with the endzones as 0-10 and 110-120 on the x-axis of Quadrant 1)
            - y: player position on "width of field" (0-53.3 yards, same thinking as "x" but this is y-axis of Quadrant 1)
            - s: speed in yard/second (numeric)
            - a: acceleration in yards/second_squared (numeric)
            - o: orientation of player (degrees)
            - a: angle of player motion (degrees)
                - 0-360 degrees, from "press box" view of field...
                    - 0 degrees facing far sidelines
                    - 90 degrees facing right endzone
                    - 180 degrees facing near sidelines
                    - 270 degrees facing left endzone
                    - 360 degrees (all the way back around)
    - ball_land_x: ball landing spot (x-axis)
    - ball_land_y: ball landing spot (y-axis)
