```python
def plot_play_action_impact(data: pd.DataFrame, output_path: str = None, show: bool = True) -> None:
    """
    Function:
    - Visualizes the pass completion rate by play action using a bar plot.

    Parameters:
    <data> (pd.DataFrame): DataFrame containing 'play_action' and 'pass_result' columns.
    <output_path> (str, optional): File path to save the plot. If None, plot is not saved.
    <show> (bool): Whether to display the plot (default: True).

    Returns:
    - None
    """
    if data.empty:
        print("No data available.")
        return
    fig, ax = plt.subplots(figsize=(8, 6))
    try:
        if 'play_action' in data.columns:
            pa_stats = data.groupby('play_action').agg({
                'pass_result': lambda x: (x == 'C').mean() * 100 if len(x) > 0 else 0
            })
            pa_values = pa_stats['pass_result'].values
            pa_labels = [f"PA={i}" for i in pa_stats.index]
            ax.bar(range(len(pa_values)), pa_values, color='#16a085')
            ax.set_xticks(range(len(pa_values)))
            ax.set_xticklabels(pa_labels)
            ax.set_ylabel('Completion %')
            ax.set_title('Play Action Impact on Completion Percentage', fontweight='bold')
            if output_path:
                plt.savefig(output_path, dpi=150, bbox_inches='tight')
            if show:
                plt.show()
        else:
            ax.text(0.5, 0.5, 'No play_action data', ha='center', va='center')
            if show:
                plt.show()
    except Exception:
        ax.text(0.5, 0.5, 'No data', ha='center', va='center')
        if show:
            plt.show()
    finally:
        plt.close()

plot_play_action_impact(input_df)
```
